```python
# check if all duplicates are most of the table if you don't include UNIQUE_INDEX + PERSON_NUMBER
def check_if_duppes(df):
    to_drop = ['UNIQUE_INDEX', 'PERSON_NUMBER']
    voi = [col for col in df.columns if col not in to_drop]
    share_dups = df.duplicated(voi).sum() / df.shape[0]
    return share_dups
```

```python
# by case?
by_case = [
    result_of_stop,
    contraband_evidence_discovered,
    type_of_property_seized,
    action_taken,
    reason_for_stop,
    basis_for_property_seizure,
    basis_for_search,
]

# by person?
by_person = [
    disability,
    gender,
    race_ethnicity,
]
```

```python
for case in by_case:
    print(check_if_duppes(case))
```

```python
for case in by_person:
    print(check_if_duppes(case))
```

```python
for case in by_case:
    case.drop(columns = ['DOJ_RECORD_ID', 'PERSON_NUMBER'], inplace = True)
```

```python
for case in by_person:
    case.drop(columns = ['DOJ_RECORD_ID', 'PERSON_NUMBER'], inplace = True)
```

```python
for case in by_case:
    case.set_index("UNIQUE_INDEX", inplace = True)

for case in by_person:
    case.set_index("UNIQUE_INDEX", inplace = True)
```
